Flatten and Ravel
Flatten --> returns the copy of array collapsed into 1D. Changes to the flatten array not affect the original array

Ravel --> Returns the view of array in collapsed 1D. Changes to raveled array affect the original array

```python
a
```

```python
bn=a.flatten() # it only make one dimnesion only and make copy not view
```

```python
bn
```

```python
bn[3]=98
```

```python
bn
```

```python
a
```

```python
bn=a.ravel() # it also make one dimension but it make view not copy. Change in view =change in original
```

```python
bn
```

```python
bn[7]=999
```
